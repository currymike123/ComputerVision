```python
#(1) 50,75,100,125,150,175,200,225,250. Show this sequence of values in a 1D and 2D array.
import numpy as np  
oneDarray = np.array([50,75,100,125,150,175,200,225,250])              #one dimentional and two dimentional array's are initialised as numpy array 
twoDarray = np.array([[50,75,100],[125,150,175],[200,225,250]])        #to utilise all the np pre-defined functions

print(f'one dimentional array :{oneDarray}')
print(f'two dimentional array :{twoDarray}')
```

```python
#(2) 50,75,100,125,150,175,200,225,250. In a 1D array, what is the array position of the value 150?
#we can get the index of any value in the numpy array with pre-defined function called np.argwhere(condition) and it will return in 2d array, as we are 
#dealing with 1d array so we use[0][0] to get the first co-ordinate of the first match.
value = 150
index = np.argwhere(oneDarray == value)[0][0]      
print(f'the array position of value 150 : {index}')
```

```python
#(3) 50,75,100,125,150,175,200,225,250. In a 2D array, what is the array position of the value 150?
#here we are dealing with 2D arrays, so we are not using [0][0]
index1 = np.argwhere(twoDarray == value)
print(f'the array position of value 150 in 2d array is {index1}')
```
