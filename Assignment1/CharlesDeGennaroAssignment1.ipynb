### 1: 50,75,100,125,150,175,200,225,250. Show this sequence of values in a 1D and 2D array.

```python
list = [x * 25 + 50 for x in range(9)]
print('Python List:', list)
array1d = np.array(list)
print('Numpy 1D Array:', array1d)
array2d = np.array([list[0:3], list[3:6], list[6:9]])
print('Numpy 2D Array:\n', array2d)
```

### 2: 50,75,100,125,150,175,200,225,250. In a 1D array, what is the array position of the value 150?

A: 4

```python
index = np.where(array1d == 150)
print('Index of 150:', index[0][0])
```

### 3: 50,75,100,125,150,175,200,225,250. In a 2D array, what is the array position of the value 150?

A: [1][1]

```python
print(f'Index of {array2d[1][1]}: [1][1]')
```
