```python
#importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating 1D and 2D arrays
onedarray = np.array([50,75,100,125,150,175,200,225,250])
twodarray = np.array([[50,75,100],[125,150,175],[200,225,250]])

print("1D array:",onedarray)
print("2D array:",twodarray)
```

```python
plt.imshow(twodarray, cmap='gray')
```

```python
#position of 150 in 1D array
res1 = np.argwhere(onedarray == 150)
print(res1)

#element in 4
print(onedarray[4])
```

```python
#position of 150 in 2D array
res2 = np.argwhere(twodarray == 150)
print(res2)

#element in 1,1
print(twodarray[1][1])
```

```python
#reading and saving an image in a variable
img = cv2.imread('messi.jpeg',1)

#converting the image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#plotting the image
plt.imshow(img)
```
