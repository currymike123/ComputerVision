#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

#### Import image

```python
img = cv2.imread('Input.png')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

#### Calc hsv

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

hist = cv2.calcHist([img_hsv], [0],None, [256], [0,255])
plt.plot(hist, color = 'red')
plt.xlim([0,255])

hist = cv2.calcHist([img_hsv], [1],None, [256], [0,255])
plt.plot(hist, color = 'green')
plt.xlim([0,255])

hist = cv2.calcHist([img_hsv], [2],None, [256], [0,255])
plt.plot(hist, color = 'blue')
plt.xlim([0,255])

plt.legend(('hue','saturation','value'), loc = 'upper left')

plt.show()
```

#### Create mask

```python
mask = cv2.inRange(img_hsv, (0, 0, 0), (30, 255, 252))

plt.imshow(mask, cmap = 'gray')
```

#### And mask to image

```python
img_masked = cv2.bitwise_and(img, img, mask = mask)

plt.imshow(img_masked)
```
