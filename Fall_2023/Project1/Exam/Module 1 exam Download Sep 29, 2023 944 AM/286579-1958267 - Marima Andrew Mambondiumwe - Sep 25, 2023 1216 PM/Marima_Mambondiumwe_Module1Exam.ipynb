```python
# Using OpenCV, NumPy and Matplotlib.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Reading my img.
img = cv2.imread('../Marima_Mambondiumwe_Module 1 Exam/Input.png')

# Converting BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Displaying the original Image
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
# Converting the Image to grayscale.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Creating a histogram.
hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

# Plotting the histogram.
plt.hist(gray_img.ravel(), 256, [0, 256])
plt.title('Histogram of the Full Image')
plt.show()
```

```python
mask = cv2.imread("Output.png", cv2.IMREAD_GRAYSCALE)
ret, mask = cv2.threshold(mask, 145, 255, cv2.THRESH_BINARY)
# Loading my Image.
img = cv2.imread('../Marima_Mambondiumwe_Module 1 Exam/Input.png')

# Converting the Image to grayscale.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Converting the Image to HSV color space.
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Applying the mask to the HSV Image.
img_hsv = cv2.bitwise_and(img_hsv, img_hsv, mask=mask)

# Analyzing the histogram to determine the ranges.
lower_part = np.array([0, 20, 70], dtype=np.uint8)
upper_part = np.array([20, 255, 255], dtype=np.uint8)

# Creating a mask based on the this range.
pic_mask = cv2.inRange(img_hsv, lower_part, upper_part)

# Applying the mask to the original Image.
masked_image = cv2.bitwise_and(img, img, mask=pic_mask)

plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
