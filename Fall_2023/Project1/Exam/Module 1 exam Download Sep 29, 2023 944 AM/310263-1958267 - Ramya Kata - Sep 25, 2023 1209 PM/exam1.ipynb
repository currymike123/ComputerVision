```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./Input.png')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

```python
img = cv2.imread('./Input.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Masked_Image=img[550:1200,0:1500]
plt.imshow(Masked_Image)
```

```python
#RGB to HSV
Masked_Image = cv2.cvtColor(Masked_Image, cv2.COLOR_RGB2HSV)

#Creating histogram
Hist_Value = cv2.calcHist([Masked_Image],[0],None,[256],[0,255])
plt.plot(Hist_Value, color='red')
# x-axis
plt.xlim([0,255])

#saturation
Hist_Value = cv2.calcHist([Masked_Image],[1],None,[256],[0,255])

plt.plot(Hist_Value, color='green')

plt.xlim([0,255])

#value
Hist_Value = cv2.calcHist([Masked_Image],[2],None,[256],[0,255])
#adding histopgram to the plot and limiting it to x-axis
plt.plot(Hist_Value, color='blue')

plt.xlim([0,255])

plt.legend(('hue','saturation','value'), loc = 'upper left')
#Image
plt.show()
```

```python
#2D
Mask=np.zeros(img.shape[:2],np.uint8)

Mask[550:1200,0:1500]=255
#confirming about grey values 
plt.imshow(Mask,cmap="gray")
```

```python
img = cv2.imread('./Input.png')
# Converting to RGB
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

Masked_Image=cv2.bitwise_and(img,img,mask=Mask)
#image Display
plt.imshow(Masked_Image)
```

```python
# HSV values
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

Hist_Value = cv2.calcHist([img],[0],Mask,[256],[0,255])
# Adding histogram
plt.plot(Hist_Value, color='red')

plt.xlim([0,255])

# Saturation
Hist_Value = cv2.calcHist([img],[1],Mask,[256],[0,255])

plt.plot(Hist_Value, color='green')

plt.xlim([0,255])

# histogram value of the image
Hist_Value = cv2.calcHist([img],[2],Mask,[256],[0,255])

plt.plot(Hist_Value, color='blue')

plt.xlim([0,255])


# Adding legend
plt.legend(('hue','saturation','value'), loc = 'upper left')

plt.show()
```

```python
img = cv2.imread('./Input.png')
#BGR to RGB
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#RGB to HSB
Img_HSV=cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

#HSV values
HSV_Minim = np.array([20, 150, 150], np.uint8)
# Maximum HSV 
HSV_Maxim = np.array([30, 255, 255], np.uint8)

#Binary mask Creation
Area_Img = cv2.inRange(Img_HSV, HSV_Minim, HSV_Maxim)

#Bitwise
HSV_Color = cv2.bitwise_and(Img_HSV, Img_HSV, mask=Area_Img)
#HSV to RGB
HSV_Color = cv2.cvtColor(HSV_Color, cv2.COLOR_HSV2RGB)
#Final Image
plt.imshow(HSV_Color)
```
