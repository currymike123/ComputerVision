```python
import cv2
import numpy as np
from numpy import interp

def rgb_to_hsv(rgb):
    rgb = rgb/255

    h = 0.0
    s = 0.0
    v = 0.0

    v = np.max(rgb)
    vMin = np.min(rgb)

    if(v > 0.0):
        s = (v - vMin) / v
    else:
        s = 0.0

    diff = (v - vMin)

    if(rgb[0] == rgb[1] and rgb[1] == rgb[2]):
        h = 0
    
    elif(rgb[0] == v):
        h = 60 * (rgb[1] - rgb[2]) / diff

    elif(rgb[1] == v):
        h = 120 + 60*(rgb[2] - rgb[0]) / diff

    elif(rgb[2] == v):
        h = 240 + 60 * (rgb[0] - rgb[1]) / diff

    h = round(h/2)
    s = round(np.interp(s, [0,1], [0,255]))
    v = round(np.interp(v, [0,1], [0,255]))
    
    return np.array([h,s,v])

rgb = np.array([200,74,55])

hsv = rgb_to_hsv(rgb)

print("RGB Value: ", rgb)
print("HSV value: ", hsv)

```

```python
import cv2
import numpy as np
from numpy import interp

def hsv_to_rgb(hsv):

    #Initialize rgb variables
    r = 0
    g = 0
    b = 0

    #Extract HSV values and map them to h[0,360], s[0,1], v[0,1]
    h = hsv[0] * 2
    s = np.interp(hsv[1], [0,255], [0,1])
    v = np.interp(hsv[2], [0,255], [0,1])

    #Convert values based on formula
    c = v * s
    x = c * (1-abs((h/60) % 2 -1))

    m = v-c
    
    if(h >= 0 and h < 60):
        r = c
        g = x
        b = 0

    elif(h >= 60 and h < 120):
        r = x
        g = c
        b = 0

    elif(h >= 120 and h < 180):
        r = x
        g = c
        b = 0

    elif(h >= 180 and h < 240):
        r = x
        g = c
        b = 0

    elif(h >= 240 and h < 300):
        r = x
        g = c
        b = 0

    elif(h >= 300 and h < 360):
        r = x
        g = c
        b = 0

    #map the values and round them to whole integers
    r = round((r + m)*255)
    g = round((g + m)*255)
    b = round((b + m)*255)

    #Return the RGB values as an array
    return np.array([r,g,b])


hsv = np.array([4, 185, 200])

rgb = hsv_to_rgb(hsv)

print("HSV value: ", hsv)
print("RGB Value: ", rgb)


    
```
