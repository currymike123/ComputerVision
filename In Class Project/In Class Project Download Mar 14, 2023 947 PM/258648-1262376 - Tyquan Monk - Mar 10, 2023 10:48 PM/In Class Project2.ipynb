Nested Loop

White Pixels 

```python
h = img.shape[0]
w = img.shape[1]

white = np.array([255,255,255])
black = np.array([0,0,0])
count = 0
for y in range(0,h):
    for x in range(0,w):
        if np.array_equal(img[y,x],white):
           print("Pixel [",y,"]","[",x,"]", "is the color white.") 
count+=1
print("There are",count,"white pixels.")

count = 0

for y in range(0,h):
    for x in range(0,w):
        if np.array_equal(img[y,x],black):
            print("Pixel [",y,"]","[",x,"]", "is the color white.")
count+=1
print("There are",count,"white pixels.")
```

Black Pixels 

```python

count = 0
for y in range(0,h):
    for x in range(0,w):
        if np.array_equal(img[y,x],black):
            print("Pixel [",y,"]","[",x,"]", "is the color black.")
            count+=1

print("There are",count,"black pixels.")
```

Red Pixels 

```python
imgRed = copy.deepcopy(img)

red = np.array([255,0,0])

for y in range(0,h):
    for x in range(0,w):
        if np.array_equal(imgRed[y,x],black):
            imgRed[y,x] = red

plt.imshow(imgRed)
```

GreyScale

```python

imgGrey = copy.deepcopy(img)

for y in range(0,h):
    for x in range(0,w):
        imgGrey[y,x,0] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3
        imgGrey[y,x,1] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3
        imgGrey[y,x,2] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3

#Combine the greyscale and the original image so they can be displayed at the same time. axis=0 for vertica
imgTwo = np.concatenate((img, imgGrey), axis=1)
#Draw the image. Use matplotlib while using a python notebook.
plt.imshow(imgTwo)
```

Threshold Filter Image

```python
imgGrey = copy.deepcopy(img)
imgThres = np.zeros((200,200,3), dtype=np.uint8)

for y in range(0,h):
    for x in range(0,w):
        imgGrey[y,x,0] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3
        imgGrey[y,x,1] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3
        imgGrey[y,x,2] = img[y,x,0]/3 + img[y,x,1]/3 + img[y,x,2]/3
        if (imgGrey[y,x,0] > 100):
            imgThres[y,x] = white
        else:
            imgThres[y,x] = black

imgTwo = np.concatenate((imgGrey, imgThres), axis=1)

plt.imshow(imgTwo)
```

```python
import numpy as np
from numpy import interp
def rgb_to_hsv(rgb):

    rgb = rgb/255

    h = 0.0
    s = 0.0
    v = 0.0
# Find the max and min RGB values.
    v = np.max(rgb)
    vMin = np.min(rgb)
# Set the saturation value.
    if(v>0.0):
        s = (v - vMin)/v
    else:
        s = 0.0
# Calculate (v - vMin) convenience
    diff = (v - vMin)
# Compute the hue by the relative sizes of the RGB components
# Are r,g,b equal.
    if(rgb[0,0,0] == rgb[0,0,1] and rgb[0,0,1] == rgb[0,0,2]):
        h = 0
# Is the point within +/- 60 degrees of the red axis
    elif(rgb[0,0,0] == v):
        h = 60 * (rgb[0,0,1] - rgb[0,0,2]) / diff
# Is the point within +/- 60 degrees of the green axis
    elif(rgb[0,0,1] == v):
        h = 120 + 60 * (rgb[0,0,2] - rgb[0,0,0]) / diff
# IS the point within +/- 60 degrees of the blue axis
    elif(rgb[0,0,2] == v):
        h = 240 + 60 * (rgb[0,0,0] - rgb[0,0,1]) / diff
# interp function is used to convert the value of one range to other range
s = round(np.interp(s,[0,1],[0,255]))
v = round(np.interp(v,[0,1],[0,255]))
h=round(h/2)
# Return hsv values.
return np.array([[[h,s,v]]])
```
