Digital images are made up of a grid (2D Array) of pixels.  Each location in the grid is given three values.  The combonation of these values reprents the pixel color. 

Color can be interpreted in many different ways.  The two main color systems we will discuss are: 

#### **RGB:** <span style="color:red">Red</span>,<span style="color:green"> Green</span>, <span style="color:blue">Blue</span>

A list of three 8-bit integers is the most common way to represent RGB values.

8-bit: 256 values from 0-255.

+ (255,0,0) = Red
+ (0,255,0) = Green
+ (0,0,255) = Blue
+ (0,0,0) = Black
+ (255,255,255) = White

(16, 24, or 36... bit color can be used for higher fidelity.) 

#### **HSB:** Hue, Saturation, Brightness
