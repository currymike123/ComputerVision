### Color

Our perception of color comes from our ability to detect a small range of the electromagnetic spectrum which we call the visible spectrum or more commonly visible light. 

<img src="Graphics/Linear_visible_spectrum.png">

There are two ways to interpert color: **Additive** (RGB) or **Subtractive** (CMYK).  

+ **Additive** color is produced by adding together wavelengths of light.  Additive color starts at black (absence of light) and moves to white as more light is added.  

+ **Subtractive** color is produced by removing, partially or totally, certain wavelengths of light while allowing the remaining wavelengths through.  Subtractive color starts at white (pure light) and moves to black as wavelengths are filtered out.   

### Digital Images 

Digital color is most commonly represnted in the **RGB** (additive) color system.  If you are reading this on a computer monitor, tablet, or  smart phone, then it is emmiting wavelengths of light that your eyes are perceiving as color.  To make a brighter color the device emits more light.  What happens when you turn your device off?  It returns to black.  The magic of additive color!     

Digital images are made up of a grid of pixels each storing a color value.  When programming this grid can be represented in a 2D array (go into detail here).  Each location in the grid is given three values.  The combonation of these values reprents the pixel color value. 

#### **RGB:** <span style="color:red">Red</span>,<span style="color:green"> Green</span>, <span style="color:blue">Blue</span>

A list of three 8-bit integers is the most common way to represent RGB values.

8-bit: 256 values from 0-255.

+ (255,0,0) = Red
+ (0,255,0) = Green
+ (0,0,255) = Blue
+ (0,0,0) = Black
+ (255,255,255) = White

256 x 256 x 256 = 16,77,216 colors can be represented 

(16, 24, or 36... bit color can be used for higher fidelity.)
