## Common File Types

### **Raster**  (Pixel Array)

The raster file formats all store image data in an array structure.  Different formats allow for certain types of compression, color spaces, and bit depth.  Separating formats into lossy and lossless.

+ **TIFF** Tag Image File Format. (1986)  Orginated by the Aldus Corp and later bought by Adobe.  Tiff or Tif is very general and complex format.  Cross platform compatible.  Can store both lossy (Jpeg) and lossless image data.
[TIFF Wiki](https://en.wikipedia.org/wiki/TIFF) <br><br>
+ **JPEG**  (JPG) Joint Photographic Experts Group. (1992)  Jpeg's main advantage is high-quality practical compression. [JPEG Wiki](https://en.wikipedia.org/wiki/JPEG)<br><br>
+ **GIF**  The General Graphics Interchange Format. (1987)  Color must be represented in 8-bits total for all color channels so limited to 256 colors. 16-bit option also available but the 8-bit is most popular.  The limited file size and ability to store multiple images in one file made the format popular in the early days of the internet to today. [GIF Wiki](https://en.wikipedia.org/wiki/GIF) <br><br>
+ **PNG** Portable Network Graphics. (1996) PNG Working Group and donated to W3. Designed as a replacement for the GIF.  Lossless compression RGB 24-bit color. 

### **Vector** 

The vector files formats are built using coded lines and angles to draw images, type, graphics.  

+ **Postscript** Page Description Language. (1982)  Invented by Adobe.  PDF, EPS.  Dominates the graphics and printing industry. [Postscript Wiki](https://en.wikipedia.org/wiki/PostScript)<br><br>
+ **SVG**  Scalable Vector Graphics. (1999) World Wide Web Consortium (W3C)  Vector graphics stored in XML files which allows easy searching.  Allows for multiple files and animation. [SVG Wiki](https://en.wikipedia.org/wiki/Scalable_Vector_Graphics)
