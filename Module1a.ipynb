### Three Axis Color

RGB & CMYK can be represented in a cube.<br>

Other color spaces exist such as HSV (HSB) & YCrCb.  HSV stands for Hue, Saturation, and Value.  You may see it called HSB too (brightness instead of value).  You can visualize the HSV color space in a cylinder. 

<img src="Graphics/rgbvshsb.jpeg"><br>

YCrCb, YCbCr, ...

Is primary a broadcast color space.  It's main advantage is in compression and subsampling.  Will be discussed more in Image Compression Module 3a<br>
+  The Y stands for the Luma value (Brightness) or the image.  <br>
+ Cr stands for the change in the red channel.<br>
+ Cb stands for the change in the blue channel. <br>
