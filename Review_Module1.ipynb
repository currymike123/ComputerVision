```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read the image. 1 = Color. 0 = B/W.
img = cv2.imread('Graphics/face.png', 1)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Print the shape of the image
print("Original Image:", img.shape)

adjust_img = img[75:100, 125:175, :]

print("Sliced Image", adjust_img.shape)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(adjust_img)
plt.title("Sliced Image")

plt.show()
```
