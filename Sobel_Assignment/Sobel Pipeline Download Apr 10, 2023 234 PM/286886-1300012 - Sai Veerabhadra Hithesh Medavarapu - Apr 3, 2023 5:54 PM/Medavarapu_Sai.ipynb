```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# importing image 
img_preblur = cv2.imread('messi.jpeg')

# Gray-Scale conversion
img_preblur = cv2.cvtColor(img_preblur, cv2.COLOR_BGR2GRAY)

plt.imshow(img_preblur, cmap='gray')
```

```python
#noise reduction
img = cv2.GaussianBlur(img_preblur,(5,5),0)

plt.imshow(img, cmap='gray')
```

```python
# Sobel filtering 
Sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, 3)
Sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, 3)
```

```python
# Magnitude calculation
Magnitude = cv2.magnitude(Sobelx,Sobely)

plt.imshow(Magnitude, cmap="gray")
```

```python
# Applying threshold 
th, image = cv2.threshold(Magnitude, 70, 255, cv2.THRESH_BINARY)

# Displaying 
fig = plt.figure(figsize= (10,10))

fig.add_subplot(1,2,1)
plt.imshow(img_preblur,cmap='gray')
plt.title('Original')

fig.add_subplot(1,2,2)
plt.imshow(image,cmap='gray')
plt.title('Sobel')
```
