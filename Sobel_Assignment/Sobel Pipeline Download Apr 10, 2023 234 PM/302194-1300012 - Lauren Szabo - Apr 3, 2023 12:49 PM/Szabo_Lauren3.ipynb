```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("images/CSC_0078.jpg",cv2.IMREAD_GRAYSCALE)


plt.imshow(img,cmap='gray')
```

```python
gaussian_filter = np.array([[1,2,1],[2,4,2],[1,2,1]], dtype=np.float32) / 16

imgGaussian = cv2.GaussianBlur(img, (3,3),0)

fig = plt.figure(figsize = (20,20))

fig.add_subplot(1,2,1)

plt.imshow(img,cmap='gray')
plt.title = ('original Image')

fig.add_subplot(1,2,2)

plt.imshow(imgGaussian,cmap='gray')
plt.title = ('filtered Image')# I feel like this didnt blur that much? but I dont see anything wrong with the code
```

```python
sobelx = cv2.Sobel(imgGaussian,cv2.CV_32F,1,0,ksize=3)
sobely = cv2.Sobel(imgGaussian,cv2.CV_32F,0,1,ksize=3)

magnitude = cv2.magnitude(sobelx,sobely)
img_thresh = cv2.threshold(magnitude, 100, 255, cv2.THRESH_BINARY)

plt.imshow(img_thresh[1],cmap='gray')# it was not working without specifying the thresholds img array (apparently its a tuple??)
```
