```python
# import and show image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Import image
img = cv2.imread('siri.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
#Noise reduction using Gaussian Blur
blur_img = cv2.GaussianBlur(img, (9,9), 0)
plt.imshow(blur_img, cmap='gray')
```

```python
# Applying Sobel operators to calculate horizontal and vertical gradients
Sobel_Gradx = cv2.Sobel(blur_img, cv2.CV_64F, 1, 0)
Sobel_Grady = cv2.Sobel(blur_img, cv2.CV_64F, 0, 1)

sobelxy = cv2.add(Sobel_Gradx,Sobel_Grady)
plt.imshow(sobelxy, cmap="gray")
```

```python
# Calculating the magnitude of gradient using cv2.magnitude
magnitude = cv2.magnitude(Sobel_Gradx, Sobel_Grady)

plt.imshow(magnitude, cmap='gray')
```

```python

# Applying the threshold to the magnitude image to get a binary edge map
retval, img_edge = cv2.threshold(magnitude, 50, 255, cv2.THRESH_BINARY)

plt.imshow(img_edge, cmap='gray')
```
