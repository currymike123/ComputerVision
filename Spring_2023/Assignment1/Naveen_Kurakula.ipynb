```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#one dimensional array
onedem = [50,75,100,125,150,175,200,225,250]
#two dimensional array
twodem = [[50,75,100],[125,150,175],[200,225,250]]
#print one dimensional array
print(onedem)
#print two dimensional array
print(twodem)
```

```python
#one dimensional array
onedem = [50,75,100,125,150,175,200,225,250]
#Print the 150 array position in one dimensional array
print(onedem.index(150))
```

```python
#two dimensional array
twodem = np.array([[50,75,100],[125,150,175],[200,225,250]])
#Print the 150 array position in two dimensional array
fun = np.argwhere(twodem == 150)
print(fun)
```
