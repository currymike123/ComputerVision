### Question 3

```python
#Function to fetch the index of an element in a 2-D array
def index_2d(myList, v):
    for i, x in enumerate(myList):
        if v in x:
            return [i, x.index(v)]
```

```python
# reference link for the above method
#https://stackoverflow.com/questions/5775352/python-return-2-ints-for-index-in-2d-lists-given-item
```

```python
index_2d(array_2D, 150) #using the above function to fetch the index value of the 150 in the 2-D array
```
