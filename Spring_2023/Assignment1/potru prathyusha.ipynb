```python
#import libraries
import numpy as np
# Create a list with the seqence of values
l= 50,75,100,125,150,175,200,225,250
# create an 2D-array using list
array2 = np.array(l).reshape(3,3)
#print 2D-array
print(array2)

value = 150
#get position in 2D-array
for i in range(3):
  for j in range(3):
    if value == array2[i][j]:
        position = (i, j)
        break
#print position
print("The position of value", value, "in the 2D array is", position)
```
