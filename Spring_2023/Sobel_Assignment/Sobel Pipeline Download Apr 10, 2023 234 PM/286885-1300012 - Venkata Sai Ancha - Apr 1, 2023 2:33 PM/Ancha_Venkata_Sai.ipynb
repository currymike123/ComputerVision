# Step 1 ->  Grayscale conversion

```python
# importing image and converting the BGR image to Gray image

import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/Users/venkatasaiancha/Documents/sem 2/image processing/photos/gadot.jpeg")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
```

# Step 2 ->  Noise reduction

```python
# apply a Gaussian blur with kernel size (5,5) and standard deviation 0
img_blurred = cv2.GaussianBlur(img_gray, (5,5), 0)
plt.imshow(img_blurred, cmap='gray')
```

```python
# resized the image ,as guassian blur was not working for the big image
resizeImg = cv2.resize(img_gray, None, fx= 0.5, fy= 0.5, interpolation= cv2.INTER_LINEAR)
plt.imshow(resizeImg, cmap="gray")
```

```python
# again appling a Gaussian blur with kernel size (5,5) and standard deviation 0 to resized blurred image.
resized_blurred = cv2.GaussianBlur(resizeImg, (5,5), 0)
plt.imshow(resized_blurred, cmap='gray')
```

# Step 3 -> Horizontal and vertical Sobel filtering

```python
# Apply horizontal and vertical Sobel filters using cv2.Sobel()
sobelx = cv2.Sobel(resized_blurred, cv2.CV_64F, 1, 0, ksize=3)
```

```python
sobely = cv2.Sobel(resized_blurred, cv2.CV_64F, 0, 1, ksize=3)
```

```python
# Plotting the sobelx 
plt.imshow(sobelx, cmap="gray")
```

```python
# Plotting the sobely 
plt.imshow(sobely, cmap="gray")
```

```python
# combined image of sobelx and sobely is plotted by using cv2.add function
complet_sobel = cv2.add(sobelx,sobely)
plt.imshow(complet_sobel, cmap="gray")
```

# Step 4 -> Magnitude calculation

```python
# Calculating the magnitude of gradient using cv2.magnitude()
magnitude = cv2.magnitude(sobelx, sobely)
```

```python
plt.imshow(magnitude, cmap="gray")
```

# Step 5 -> Thresholding

```python
# Apply thresholding using cv2.threshold()
thresh_value = 100
max_value = 255
thresh_type = cv2.THRESH_BINARY
retval, thresholded = cv2.threshold(magnitude, thresh_value, max_value, thresh_type)
```

```python
plt.imshow(thresholded, cmap='gray')
```
