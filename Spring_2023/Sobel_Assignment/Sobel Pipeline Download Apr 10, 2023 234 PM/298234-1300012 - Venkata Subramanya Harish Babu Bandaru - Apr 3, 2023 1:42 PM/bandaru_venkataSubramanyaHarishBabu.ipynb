```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# 1. Read and convert the image to Gray scale

```python
img_org = cv2.imread('C:/Users/HP/Downloads/kaj.jpg')
img_gray = cv2.cvtColor(img_org, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

# 2. Noise reduction using guassian blur

```python
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0) 
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

# 3. Sobel filter applied horizontally, vertically and both

```python
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
```

```python
plt.imshow(sobelx, cmap = 'gray')
```

```python
plt.imshow(sobely, cmap = 'gray')
```

```python
plt.imshow(sobelxy, cmap = 'gray')
```

# 4. calculating magnitude of the image using sobel x and sobel y

```python
img_mag = cv2.magnitude(sobelx, sobely)
plt.imshow(img_mag, cmap = 'gray')
```

## tried sobelxy image to get magnitudes

```python
img_mag_2 = cv2.magnitude(sobelxy, np.zeros(sobelxy.shape))
plt.imshow(img_mag, cmap = 'gray')
```

# 5. Thresholding

```python
img_thres = cv2.threshold(img_mag, 150, 225, cv2.THRESH_BINARY) #Applied some random threshold values to get the initial image
```

```python
plt.imshow(img_thres[1], cmap= 'gray')
```
