```python
#Import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./rachel.jpg")

#convert to grayscale
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap='gray')

```

```python
#convert to blur image
blurred_img = cv2.GaussianBlur(image, (5,5), 0)
plt.imshow(blurred_img, cmap='gray')
```

```python
sobelx = cv2.Sobel(blurred_img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred_img, cv2.CV_64F, 0, 1, ksize=3)
sobelxy = cv2.add(sobelx,sobely)
plt.imshow(sobelxy, cmap="gray")
```

```python
magnitude = cv2.magnitude(sobelx, sobely)
plt.imshow(magnitude, cmap="gray")
```

```python
tshld_value = 155
max_value = 240
tshld_type = cv2.THRESH_BINARY
retval, thresholded = cv2.threshold(magnitude, tshld_value, max_value, tshld_type)
plt.imshow(thresholded, cmap='gray')
```
