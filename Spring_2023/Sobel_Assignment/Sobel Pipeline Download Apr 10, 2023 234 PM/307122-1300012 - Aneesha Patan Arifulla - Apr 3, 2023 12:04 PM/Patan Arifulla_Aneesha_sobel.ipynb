```python

import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("C:/Users/HP/OneDrive/Desktop/Image_A/salena.webp")
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
```

```python
blurred_img = cv2.GaussianBlur(gray_image, (5,5), 0)
plt.imshow(blurred_img, cmap='gray')
```

```python
sobelx = cv2.Sobel(blurred_img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred_img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
sobelxy = cv2.add(sobelx,sobely)
plt.imshow(sobelxy, cmap="gray")
```

```python
magnitude = cv2.magnitude(sobelx, sobely)
plt.imshow(magnitude, cmap="gray")
```

```python
thresh_value = 100
max_value = 255
thresh_type = cv2.THRESH_BINARY
retval, thresholded = cv2.threshold(magnitude, thresh_value, max_value, thresh_type)
```

```python
plt.imshow(thresholded, cmap='gray')
```
