```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
img=cv2.imread("robert_pattinson.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
```

```python

imgGaussian=cv2.GaussianBlur(img,(7,7),0)

plt.imshow(imgGaussian,cmap="gray")
```

```python
 #Sobel convolution
imgSobelX = cv2.Sobel(imgGaussian, cv2.CV_64F, 1,0)
imgSobelY = cv2.Sobel(imgGaussian, cv2.CV_64F, 0,1)

```

```python
# Plot the combined image
fig = plt.figure(figsize = (15,15))

# Create subplots
# Left Image
fig.add_subplot(1,3,1)
plt.imshow(imgGaussian, cmap='gray')
plt.title('Original Image')

# Center Image
fig.add_subplot(1,3,2)
plt.imshow(imgSobelX, cmap='gray')
plt.title('Prewitt X Edge Detection')

# Right image
fig.add_subplot(1,3,3)
plt.imshow(imgSobelY, cmap='gray')
plt.title('Prewitt Y Edge Detection')
```

```python


imgSobel=cv2.magnitude(imgSobelX,imgSobelY)

plt.imshow(imgSobel,cmap="gray")
```

```python
retval,dst=cv2.threshold(imgSobel,100,255,cv2.THRESH_BINARY)
plt.imshow(dst,cmap="gray")
```
